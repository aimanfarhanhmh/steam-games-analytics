# JSON Response from SteamSpy API

### Retreiving data from API

```python
import requests
import pandas as pd

data = requests.get("https://steamspy.com/api.php?request=all")

print(type(data.json()))

df = pd.DataFrame(data.json()).T

df.to_csv('./top1000SteamGamesByOwners.csv')

print(data.json())
```

## Reading Game Info from Steam API

For example, we are parsing the game info for Counter Strike 2 (APPID 730)

```python
import json
import requests

APP_ID = 730
URL = f"https://store.steampowered.com/api/appdetails?appids={APP_ID}"

request = requests.get(url=URL)
json_request = request.json()

data = json_request[f"{APP_ID}"]["data"]

name:str = str(data['name'])
software_type:str = str(data['type'])
appid:int = APP_ID

if str(data['is_free']) == 'True':
    is_free:bool = True
else:
    is_free:bool = False

print(name)
print(software_type)
print(appid)
print(is_free)


```
