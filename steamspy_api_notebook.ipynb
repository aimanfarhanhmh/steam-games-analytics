```python
import pandas as pd
import numpy as np
import requests
import json
```

```python
API_URL = 'https://steamspy.com/api.php?request=all&page=0'

response = requests.get(API_URL)
print(response.status_code)
```

```python
data = json.loads(response.text)
print(data)
```

```python
df = pd.DataFrame(data).T.reset_index(drop=True)
df.index += 1
df.head(10)
```

```python
print(len(df))
```

```python
df.to_csv('./steamspy_top1000_games.csv', index_label='owner_rank')
```

```python
for appid in df['appid']:
    print(appid)
```

```python
with open('games_list.txt', 'w') as gamelist:
    for appid in df['appid']:
        gamelist.write(str(appid) + '\n')
```
